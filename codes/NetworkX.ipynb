```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
G=nx.DiGraph()
G.add_node('Repayment rate',pos=(3, 8))
G.add_node('Completion',  pos=(4, 13))
G.add_node('Student',pos=(1.5,10))
G.add_node('School',  pos=(1.5, 13))
G.add_node('Earnings',  pos=(1,3))
G.add_node('Academics',  pos=(4, 2))
G.add_node('Admissions',  pos=(5, 11))
G.add_node('Cost',  pos=(6, 6))
G.add_node('Aid', pos=(4.5,5))
G.add_edge('Student','Repayment rate',weight=1,label='')
G.add_edge('School','Repayment rate',weight=1,label='')
G.add_edge('Completion','Repayment rate',weight=1,label='')
G.add_edge('Earnings','Repayment rate',weight=1,label='')
G.add_edge('Academics','Repayment rate',weight=1,label='')
G.add_edge('Admissions','Repayment rate',weight=1,label='')
G.add_edge('Cost','Repayment rate',weight=1,label='')
G.add_edge('Aid','Repayment rate',weight=1,label='')

elarge=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] >0.5] # solid edge
esmall=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] <=0.5]

pos=nx.get_node_attributes(G,'pos')
nx.draw_networkx_nodes(G,pos, node_size=1000, node_color='c')

# Draw edges
nx.draw_networkx_edges(G,pos,edgelist=elarge, width=2, edge_color='black')
nx.draw_networkx_edges(G,pos,edgelist=esmall, arrows=False, width=2,
                       alpha=0.5,edge_color='black',style='dashed')
 
nx.draw_networkx_labels(G,pos,font_size=12,font_family='sans-serif')

edge_labels =dict([((u, v), d['label']) 
                   for u, v, d in G.edges(data=True)])

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
pylab.axis('off')
plt.savefig("relationship.png", format='png')
plt.show()
```
